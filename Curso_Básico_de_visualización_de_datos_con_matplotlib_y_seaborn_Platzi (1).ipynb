# Subplot

Subplot permite crear gr치ficos dentro de una gr치fica. Esto lo hace a trav칠s de una matriz de gr치ficos y se puede acceder a ellos a trav칠s de 칤ndices

```python
#creamos el set de datos
x = np.linspace(0,5,11)
y = x**2
```

Con subplot debemos pasarle 3 par치mentros. El primero son las filas que queremos, el segundo las columnas que queremos y el tercero es el 칤ndice del gr치fico que vamos a trabajar en ese momento. 

```python
plt.subplot(1, 2, 1) #comenzamos trabajando con el 칤ndice 1 
plt.plot(x, y, 'r*')# modificamos el color del gr치fico sin que nos afecte el otro gr치fico
plt.plot(y, x, 'b:') #invertimos la relaci칩n y modificamos el mismo gr치fico
plt.subplot(1, 2, 2) #continuamos con el 칤ndice 2 
plt.hist(x, color='m')
plt.show()
```

# M칠todo orientado a objeto

Hay distintas maneras de hacer gr치ficas dentro de Matplotlib, ya vimos pyplot; sin embargo, es muy complicado personalizarla y entrar a ciertos par치metros.

M칠todo orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gr치 ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gr치ficas las cuales contienen diferentes ejes.

Es un poco m치s complicado, pero en el mismo gr치fico podemos personalizarlo mucho mejor.

Diferencias entre Pyplot y Object Oriented

Pyplot |	Object Oriented

R치pido |	Mayor personalizaci칩n

F치cil |	M치s c칩digo

Una sola figura |	M치s amigable a m칰ltiples diagramas

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#graficamos con metodo orientado a objetos
fig = plt.figure()  #creamos una figura, es el lienzo donde vamos a dejar nuestras graficas y lo guardamos en una variable
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) #las graficas se van a llamar axes y le agregamos un axes
                                          #los dos primeros par치mentros hacen referencia a la posici칩n en la figura o en el lienzo 
                                          #los otros dos hacen referencia al tama침o que va a tener esa figura 
axes.plot(x,y, 'b')
fig.show()
```

```python
#ejemplo superponiendose
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) 
axes2 = fig.add_axes([0.1, 0.1, 0.4, 0.3])

axes.plot(x,y, 'b')
axes2.plot(y, x, 'r')
fig.show()
```

```python
#cambiando la posici칩n vemos que se logra correr un poco la figura 2
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) 
axes2 = fig.add_axes([0.3, 0.3, 0.4, 0.3])

axes.plot(x,y, 'b')
axes2.plot(y, x, 'r')
fig.show()
```

```python
#lo ponemos correctamente para que figuren las dos graficas
fig = plt.figure() 
axes = fig.add_axes([1, 0.1, 1, 1]) 
axes2 = fig.add_axes([2, 0.1, 1, 1])

axes.plot(x,y, 'b')
axes2.plot(y, x, 'r')
axes.set_title('primer grafico')
axes2.set_title('segundo grafico')
axes.set_facecolor('gray')
fig.show()
```

# Subplots

```python
x = np.linspace(0, 5, 11)
```

```python
y = np.sin(x)
```

```python
#graficamos usando subplots
fig, ax = plt.subplots()
ax.plot(x, y, 'b')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2) #establecemos que queremos una fila y dos columnas
ax[0].plot(x, y, 'b') #podemos acceder a cada ax a traves de su 칤ndice como un array 
ax[1].plot(y, x, 'r')
fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) # creamos los subplots como una tupla y accedemos a ellos como una tupla
ax1.plot(x, y, 'b') 
ax1.set_title('primer plot')
ax2.plot(y, x, 'r')
fig.show()
```

```python

fig, ax = plt.subplots(nrows=2, ncols=4)
ax[0,0].plot(x,np.cos(x))
ax[0,1].plot(x,np.sin(x),'r')
ax[0,2].plot(x,np.tan(x), 'y')
ax[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x,np.cos(x))
ax3.plot(x,np.sin(x),'r')
ax5.plot(x,np.tan(x), 'y')
ax7.plot(x,np.cos(x)**2)
fig.tight_layout() #hacer que la figura se vea m치s acomodada
```

Reto : Graficar y completar todos los ax de la figura de distinta forma. 

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x,np.cos(x), 'lime', marker='P')
ax2.plot(x,np.sin(x),'r', marker='>')
ax3.plot(x,np.tan(x), 'y', ls='-.')
ax4.plot(x,np.cos(x)**2, 'chocolate', marker='D')
ax5.plot(x**2, y, 'aqua')
ax6.plot(x**3,y, 'magenta', marker='.')
ax7.plot((x*(-x)), 'darkgray', marker='*', ls=':')
ax8.plot(x, y,'peru')
fig.tight_layout()
```

# Heatmap

Heatmap es un tipo de gr치fico enfocado a una estructura matricial. Heatmap correlaciona todas las variables num칠ricas del dataset.

Se pueden agregar diferentes par치metros. annot muestra el valor de la correlaci칩n,
cmap color,linewidthsespacio entre variables, linecolor color de las l칤neas, vminv, max valores m치ximos y m칤nimos, cbar=False eliminar la barra

```python
tips.corr() #Muestra las variables correlacionadas entre si
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=0, vmax=1, cbar=True)
plt.show()
#linewidths separa los cuadros del heatmap
#vmax y vmin nos permite modificar la escala de valores que tenemos 
#cbar=true o false permite agregar o quitar la barra
```

Documentaci칩n de heatmap: https://seaborn.pydata.org/generated/seaborn.heatmap.html
